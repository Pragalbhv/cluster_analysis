```python
def coordination_per_atom_list_voronoi_freud(atoms_list, at_list=None, min_vol=0.0):
    import numpy as np
    from collections import defaultdict
    from tqdm import tqdm
    import freud

    # Build species list
    if at_list is None:
        all_species = sorted({s for at in atoms_list for s in at.get_chemical_symbols()})
    else:
        all_species = list(at_list)
    coord_data = {sp: {nb: [] for nb in all_species} for sp in all_species}

    for atoms in tqdm(atoms_list):        
        L = atoms.get_cell()
        assert atoms.pbc.all(), "freud Voronoi expects PBC; set atoms.pbc=True in all directions."
        box = freud.box.Box.from_matrix(L.array)  # 3x3 matrix
        positions = atoms.get_positions()

        v = freud.locality.Voronoi()
        v.compute((box, positions))

        # v.polytopes lists faces; v.nlist gives neighbor pairs for shared faces
        # Filter by face area if desired (min_area)
        # Build adjacency from nlist:
        # Each neighbor pair appears once; map to species and count
        species = atoms.get_chemical_symbols()
        # face areas correspond to v.face_areas (aligned with neighbor pairs in v.nlist)
        # vols = v.volumes  # length = len(v.nlist)
        # return v

        # Replace volumes with facet areas (per-pair)
        xx = v.nlist[:,0]
        yy = v.nlist[:,1]
        areas = v.nlist.weights 
        pair_iter = zip(xx, yy, areas)
        pair_iter=sorted(pair_iter, key=lambda x: x[0])
    

        # Start per-atom counters for this frame
        per_atom_counts = [defaultdict(int) for _ in range(len(atoms))]

        unique_is=[]


        for i, j, a in pair_iter:
            # print(a)
            if a >= min_vol:
                if i not in unique_is:
                    if len(unique_is)>0:
                        central_type=species[unique_is[-1]] #previous unique i is done
                        for nb_type in all_species:#appends to list
                            coord_data[central_type][nb_type].append(per_atom_counts[unique_is[-1]][nb_type])
                            # print(central_type,nb_type)
                    unique_is.append(i)
                #update current i's register
                per_atom_counts[i][species[j]] += 1
            ###final i
            unique_is.append(i)
            central_type=species[unique_is[-1]] #previous unique i is done
            for nb_type in all_species:#appends to list
                coord_data[central_type][nb_type].append(per_atom_counts[unique_is[-1]][nb_type])

        
                

    return coord_data
```

```python
def coordination_per_atom_list_voronoi_freud2(atoms_list, at_list=None, min_area=0.0):
    import numpy as np
    from collections import defaultdict
    from tqdm import tqdm
    import freud

    # Build species universe
    if at_list is None:
        all_species = sorted({s for at in atoms_list for s in at.get_chemical_symbols()})
    else:
        all_species = list(at_list)
    sp2idx = {sp: i for i, sp in enumerate(all_species)}

    # Output structure: for each central species, a list of neighbor counts per neighbor species
    coord_data = {sp_c: {sp_n: [] for sp_n in all_species} for sp_c in all_species}

    for atoms in tqdm(atoms_list):
        assert atoms.pbc.all(), "freud Voronoi expects PBC in all directions."
        box = freud.box.Box.from_matrix(atoms.get_cell().array)
        pos = atoms.get_positions()
        species = np.array(atoms.get_chemical_symbols())
        species_idx = np.array([sp2idx[s] for s in species], dtype=int)

        v = freud.locality.Voronoi()
        v.compute((box, pos))

        # Neighbor list: i (central), j (neighbor), weight = facet area
        # The attributes below match common freud builds; if yours differs,
        # adjust to v.nlist[:,0], v.nlist[:,1], v.nlist.weights as in your code.
        i = v.nlist[:, 0]
        j = v.nlist[:, 1]
        areas = v.nlist.weights

        # Optional facet-area threshold
        mask = areas >= float(min_area)
        i = i[mask]
        j = j[mask]

        # Accumulate counts: rows=central atom index, cols=neighbor-species index
        N = len(atoms)
        S = len(all_species)
        counts = np.zeros((N, S), dtype=int)
        np.add.at(counts, (i, species_idx[j]), 1)

        # Split by central species and extend lists
        for sp_c, c_idx in sp2idx.items():
            rows = np.where(species_idx == c_idx)[0]
            if rows.size == 0:
                continue
            sub = counts[rows]  # shape: (#atoms of central species, S)
            for sp_n, n_idx in sp2idx.items():
                coord_data[sp_c][sp_n].extend(sub[:, n_idx].tolist())

        # print(counts)
        break

    return coord_data
```

```python
from ovito.io import import_file
from ovito.modifiers import CoordinationAnalysisModifier
import glob
import numpy as np
import re
from pathlib import Path

from tqdm.auto import tqdm



def calculate_rdfs(pipeline, nsamples = 500):
    nframes = pipeline.source.num_frames

    rdf_data = {}

    for frame in tqdm(range(nframes - nsamples, nframes)):
        data = pipeline.compute(frame)
        # Access the output DataTable:
        rdf_table = data.tables['coordination-rdf']

        # The y-property of the data points of the DataTable is now a vectorial property.
        # Each vector component represents one partial RDF.
        rdf_names = rdf_table.y.component_names

        # Print a list of partial g(r) functions.
        for component, name in enumerate(rdf_names):
            # print("g(r) for pair-wise type combination %s:" % name)
            
            # a, b = name.split('-')
            
            elements = sorted(name.split('-'))
            
            unique_name = '-'.join(elements)
            # unique_name = name

            y = rdf_table.y[:,component]
            # print(rdf_table.y[:,component])
            if rdf_data.get(unique_name, None) is None:
                rdf_data[unique_name] = [y]
            else:
                rdf_data[unique_name].append(y)


        # The DataTable.xy() method yields everthing as one combined NumPy table.
        # This includes the 'r' values in the first array column, followed by the
        # tabulated g(r) partial functions in the remaining columns. 

    for name in rdf_data:
        rdf_data[name] = np.vstack(rdf_data[name]).mean(axis=0)

    rdf_data["r"] = rdf_table.xy()[:, 0]
    
    return rdf_data
```

```python
def parse_composition_to_x(elements_str):
    """Calculate Pu fraction x from composition string."""
    matches = re.findall(r'([A-Z][a-z]*)(\d*)', elements_str)
    
    na_count = 0
    pu_count = 0

    for element, count_str in matches:
        count = int(count_str) if count_str else 1
        
        if element == 'Na':
            na_count += count
        elif element == 'Pu':
            pu_count += count

    total = na_count + pu_count
    return pu_count / total if total > 0 else 0.0
```

```python
import re

def extract_info(path):
    match = re.search(r'NaCl-PuCl3/x(\d*\.?\d+)/T(\d+)K/', path)
    if match:
        x = float(match.group(1))
        temperature = int(match.group(2))
        return x, temperature
    else:
        return None, None
```

```python
from collections import defaultdict

rdf_dict=defaultdict(dict)
```
